```python
import numpy as np
import tensorflow as tf
```

```python
w = tf.Variable(3., name='w')
b = tf.Variable(4., name='b')

x_1 = np.linspace(0, 100, 100)
x_2 = np.linspace(0, 100, 100)

f = lambda a, b: a * np.sin(b/10.)

# plot gradient from f
pool = []
for i in x_1:
    _pool = []
    for j in x_2:
        _pool.append(f(i, j))
    pool.append(_pool)

import matplotlib.pyplot as plt

# plot 3d gradient
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x_1, x_2, np.array(pool), cmap=plt.cm.coolwarm)

```
